```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
diabetes = pd.read_csv('/Users/artemmoskvin/Documents/Data-Science/Data/diabetes_data.csv')
diabetes.tail()
```

```python
mask = diabetes.duplicated(subset=None)
diabetes_dublicates = diabetes[mask]
print(f'Число найденных дубликатов: {diabetes_dublicates.shape[0]}')
diabetes_dedupped = diabetes.drop_duplicates(subset=None)
print(f'Результирующее число записей: {diabetes_dedupped.shape[0]}')
```

```python
display(diabetes_dedupped)
```

```python
#список неинформативных признаков
diabetes_low_information_cols = [] 
#цикл по всем столбцам
for col in diabetes_dedupped.columns:
    #наибольшая относительная частота в признаке
    top_freq = diabetes_dedupped[col].value_counts(normalize=True).max()
    #доля уникальных значений от размера признака
    nunique_ratio = diabetes_dedupped[col].nunique() / diabetes_dedupped[col].count()
    # сравниваем наибольшую частоту с порогом
    if top_freq > 0.95:
        diabetes_low_information_cols.append(col)
        print(f'{col}: {round(top_freq*100, 2)}% одинаковых значений')
    # сравниваем долю уникальных значений с порогом
    if nunique_ratio > 0.95:
        diabetes_low_information_cols.append(col)
        print(f'{col}: {round(nunique_ratio*100, 2)}% уникальных значений')
```

```python
g_diabetes = diabetes_dedupped.drop(diabetes_low_information_cols, axis=1)
print(f'Результирующее число признаков: {g_diabetes.shape[1]}')
display(g_diabetes)
```

```python
display(g_diabetes)
```

```python
g_diabetes['Insulin'] = g_diabetes['Insulin'].replace(0, float('NaN'))
g_diabetes['Glucose'] = g_diabetes['Glucose'].replace(0, float('NaN'))
g_diabetes['BloodPressure'] = g_diabetes['BloodPressure'].replace(0, float('NaN'))
g_diabetes['SkinThickness'] = g_diabetes['SkinThickness'].replace(0, float('NaN'))
g_diabetes['BMI'] = g_diabetes['BMI'].replace(0, float('NaN'))
g_diabetes.tail()
```

```python
cols_null_percent = g_diabetes.isnull().mean()
cols_with_null = cols_null_percent[cols_null_percent>0].sort_values(ascending=False)
display(cols_with_null)
```

```python
thresh = g_diabetes.shape[0]*0.7
#удаляем столбцы, в которых более 30% (100-70) пропусков
gg_diabetes = g_diabetes.dropna(how='any', thresh=thresh, axis=1)
gg_diabetes.isnull().mean()
print(gg_diabetes.shape)
```
