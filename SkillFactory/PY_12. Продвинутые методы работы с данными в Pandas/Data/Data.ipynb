```python
import pandas as pd
melb_df = pd.read_csv('/Users/artemmoskvin/Documents/Data-Science/Data/melb_data_fe.csv', sep=',')
print(melb_df)
```

```python
cols_to_exclude = ['Date', 'Rooms', 'Bedroom', 'Bathroom', 'Car'] # список столбцов, которые мы не берём во внимание
max_unique_count = 150 # задаём максимальное число уникальных категорий
for col in melb_df.columns: # цикл по именам столбцов
    if melb_df[col].nunique() < max_unique_count and col not in cols_to_exclude: # проверяем условие
        melb_df[col] = melb_df[col].astype('category') # преобразуем тип столбца
display(melb_df.info())
```

```python
melb_df.groupby(['Rooms', 'Type'])['Price'].mean()
```

```python
melb_df.groupby(['Rooms', 'Type'])['Price'].mean().unstack()
```

```python
melb_df.pivot_table(
    values='Price',
    index='Rooms',
    columns='Type',
    fill_value=0
).round()
```

```python
melb_df.pivot_table(
    values='Price',
    index='Regionname',
    columns='Weekend',
    aggfunc='count',
)
```

```python
melb_df.pivot_table(
    values='Landsize',
    index='Regionname',
    columns='Type',
    aggfunc=['median', 'mean'],
    fill_value=0
)
```

```python
melb_df.pivot_table(
    values='Price',
    index=['Method','Type'],
    columns='Regionname',
    aggfunc='median',
    fill_value=0
)
```

```python
pivot = melb_df.pivot_table(
    values='Landsize',
    index='Regionname',
    columns='Type',
    aggfunc=['median', 'mean'],
    fill_value=0
)
pivot.columns
```

```python
display(pivot['mean']['unit'])
```

```python
mask = pivot['mean']['house'] < pivot['median']['house']
filtered_pivot = pivot[mask]
display(filtered_pivot)
```

```python
import pandas as pd
ratings1 = pd.read_csv('/Users/artemmoskvin/Documents/Data-Science/Data/movies_data/ratings1.csv', sep=',')
ratings2 = pd.read_csv('/Users/artemmoskvin/Documents/Data-Science/Data/movies_data/ratings2.csv', sep=',')
dates = pd.read_csv('/Users/artemmoskvin/Documents/Data-Science/Data/movies_data/dates.csv', sep=',')
movies = pd.read_csv('/Users/artemmoskvin/Documents/Data-Science/Data/movies_data/movies.csv', sep=',')
ratings = pd.concat([ratings1, ratings2])
ratings = ratings.drop_duplicates(ignore_index=True)
ratings_dates = pd.concat([ratings, dates], axis=1)
merged = ratings_dates.merge(
    movies,
    on='movieId',
    how='left'
)
display(merged.head())

merged2 = ratings_dates.merge(
    movies,
    on='movieId',
    how='outer'
)
print('Число строк в таблице merged2: ', merged2.shape[0])
display(merged2.tail())
```

```python
merge_ratings = ratings1.merge(ratings2, how='outer')
print('Число строк в таблице merge_ratings: ', merge_ratings.shape[0])
display(merge_ratings)
```
