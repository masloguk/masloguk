```python
import pandas as pd
melb_df = pd.read_csv('/Users/artemmoskvin/Documents/Data-Science/Data/movies_data/ratings_movies.csv', sep=',')
import re 
def get_year_release(arg):
    #находим все слова по шаблону "(DDDD)"
    candidates = re.findall(r'\(\d{4}\)', arg) 
    # проверяем число вхождений
    if len(candidates) > 0:
        #если число вхождений больше 0,
	#очищаем строку от знаков "(" и ")"
        year = candidates[0].replace('(', '')
        year = year.replace(')', '')
        return int(year)
    else:
        #если год не указан, возвращаем None
        return None
melb_df['year_release'] = melb_df['title'].apply(get_year_release)
display(melb_df)
```

```python
melb_df['date'] = pd.to_datetime(melb_df['date'], dayfirst=True)
melb_df['year_rating'] = melb_df['date'].dt.year
```

```python
pivot = melb_df.pivot_table(
    values='rating',
    index='genres',
    columns='year_rating',
    fill_value=0
)
display(pivot)
display(pivot[2018])
```

```python
import pandas as pd
melb_orders = pd.read_csv('/Users/artemmoskvin/Documents/Data-Science/Data/orders_and_products/orders.csv', sep=';')
display(melb_orders)
melb_products = pd.read_csv('/Users/artemmoskvin/Documents/Data-Science/Data/orders_and_products/products.csv', sep=';')
display(melb_products)
```

```python
orders_products = melb_orders.merge(
    melb_products,
    left_on='ID товара',
    right_on='Product_ID',
    how='inner'
)
display(orders_products)
```

```python
orders_products['Profit'] = orders_products['Price'] * orders_products['Количество'] 
orders_products[orders_products['Оплачен'] == 'Да'].groupby('ID Покупателя')['Profit'].sum().sort_values(ascending=False)
```
