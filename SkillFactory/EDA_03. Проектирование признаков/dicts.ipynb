```python
import pandas as pd
import category_encoders as ce
```

```python
list_of_dicts = [
 {'product': 'Product1', 'price': 1200, 'payment_type': 'Mastercard'},
 {'product': 'Product2', 'price': 3600, 'payment_type': 'Visa'},
 {'product': 'Product3', 'price': 7500, 'payment_type': 'Amex'}
]
df = pd.DataFrame(list_of_dicts)
df
```

```python
encoder = ce.OneHotEncoder(cols=['price']) # указываем столбец для кодирования
type_bin = encoder.fit_transform(df['price'])
df01 = pd.concat([df, type_bin], axis=1)
df01
```

```python
ord_encoder = ce.OrdinalEncoder()
data_bin = ord_encoder.fit_transform(df01['payment_type'])
df02 = pd.concat([df01, data_bin], axis=1)
df02
```

```python
encoder = ce.OneHotEncoder(cols=['product','payment_type'])
cols = encoder.fit_transform(df[['product','payment_type']])
df = pd.concat([df, cols], axis=1)
df
```
