```python
import numpy as np 
import pandas as pd
import seaborn as sns
from sklearn import preprocessing
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('/Users/artemmoskvin/Documents/Data-Science/Data/wine_cleared.csv')
```

```python
data.info()
```

```python
fig, (ax1) = plt.subplots(ncols=1, figsize=(10, 8))
ax1.set_title('Исходные распределения')
sns.kdeplot(data['price'], ax=ax1, label ='beta')
```

```python
r_scaler = preprocessing.RobustScaler()
data[['price2']] = r_scaler.fit_transform(data[['price']])
fig, (ax1) = plt.subplots(ncols=1, figsize=(10, 8))
ax1.set_title('Распределения после RobustScaler')
sns.kdeplot(data['price2'], ax=ax1)
```

```python
data['price'].sum()
```

```python
s_scaler = preprocessing.StandardScaler()
data[['price2']] = s_scaler.fit_transform(data[['price']])
data = pd.DataFrame(data)
fig, (ax1) = plt.subplots(ncols=1, figsize=(10, 8))
ax1.set_title('Распределения после StandardScaler')
sns.kdeplot(data['price2'], ax=ax1)
```
